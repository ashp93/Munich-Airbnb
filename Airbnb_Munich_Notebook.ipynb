## 2.1 Data Cleaning

```python
# Renaming columns
# Note that we are using neighbourhood_cleansed instead of the neighbourhood which contain similar values but is cleaner
listings_df.rename(columns = {'neighbourhood_cleansed':'neighbourhood', 'id':'listing_id'}, inplace=True)
```

```python
# Converting prices to number (Removing $ and ,)

listings_df['price'] = listings_df['price'].str.replace('$','').str.replace(',','').astype(float)
listings_df['cleaning_fee'] = listings_df['cleaning_fee'].str.replace('$','').str.replace(',','').astype(float)

calendar_df['price'] = calendar_df['price'].str.replace('$','').str.replace(',','').astype(float)
calendar_df['adjusted_price'] = calendar_df['adjusted_price'].str.replace('$','').str.replace(',','').astype(float)
```

```python
# Parsing dates
calendar_df.head()
calendar_df['date'] = pd.to_datetime(calendar_df['date'])
reviews_df['date'] = pd.to_datetime(reviews_df['date'])

```

```python
# It seems the dates in calendar_df are wrong. There is data for 2020 which is still in the future.
# The years may have been shifted, but I don't know by how much. Therefore I will disregard the year in the analysis.
# Create month column
calendar_df['month'] = calendar_df['date'].dt.month
```

### There are outliers in the data, hence the high standard deviation.
For the sake of plotting, let's remove the outliers and plot the distribution again

```python
bins = [0, 100, 200, 300, 400, 500, 1000, 10000]
listings_df['pricebin'] = pd.cut(listings_df['price'], bins)
listings_df.groupby('pricebin').count()['listing_id']
id_price_bins = pd.DataFrame(listings_df.groupby('pricebin').count()['listing_id'])
id_price_bins['cumulative'] = id_price_bins['listing_id'].cumsum()
id_price_bins['percentage'] = id_price_bins['cumulative']/id_price_bins['listing_id'].sum()*100
id_price_bins
```

# 3. Modeling (incomplete, very low scores)

### We will drop the columns we don't need

```python
model_df = listings_df.drop([
    'latitude', 'longitude', 'price_by_nbh', 'pricebin', 'logprice',
    'pricebins_nbh', 'amenitieslist', 'num_amenities', 'amenities'], axis=1)
```

```python
model_df = pd.get_dummies(model_df, columns = model_df.select_dtypes('object').columns, drop_first=True)
```

```python
model_df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
X = model_df.drop('price', axis=1)
y = model_df['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.3, random_state =2)
```

```python
# Linear Regression
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
y_pred = linreg.predict(X_test)
y_pred_tr = linreg.predict(X_train)
```

```python
coef = pd.DataFrame()
coef['features'] = X_train.columns
coef['coef'] = linreg.coef_
coef = coef.sort_values('coef', ascending=False)
pd.concat([coef.head(),coef.tail()])
```

```python
sns.scatterplot(y_test, y_pred)
plt.xlabel('y_test')
plt.ylabel('y_pred')
plt.title("Test Error")
```

```python
sns.scatterplot(y_train, y_pred_tr)
plt.title("Training Error")
```

```python
print("Test MAE:", metrics.mean_absolute_error(y_test, y_pred))
print("Test MSE:", metrics.mean_squared_error(y_test, y_pred))
print("Test RMSE:", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('Test R2:', metrics.r2_score(y_test, y_pred))
```

```python

print("Training MAE:", metrics.mean_absolute_error(y_train, y_pred_tr))
print("Training MSE:", metrics.mean_squared_error(y_train, y_pred_tr))
print("Training RMSE:", np.sqrt(metrics.mean_squared_error(y_train, y_pred_tr)))
print('Training R2:', metrics.r2_score(y_train, y_pred_tr))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import GradientBoostingRegressor
```
